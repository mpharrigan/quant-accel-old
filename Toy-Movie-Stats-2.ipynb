Fitting the gold standard
--------
The first implied timescale at the conclusion of a long, non-adaptive-sampled simulation is used as the gold standard. *'Error'* is defined as the difference with this gold standard value. We fit the error with a single exponential.

```python
gold_popt = plot_and_fit(c.gold, p0=(1,1e3))
pp.xlabel('Walltime')
pp.ylabel('Error')
pp.title("'Gold' single long trajectory error convergence.")
pp.show()
```

We also fit exponentials to the adaptive sampling methods. In the first case, we varied the length per trajectory (*lpt*). In the second case, we varried the number of parallel (*ll*) trajectories per round.

```python
for toy in c.lpt_list[:-1]:
    plot_and_fit(toy, p0=(10,3))
pp.xlabel('Walltime')
pp.ylabel('Error')
pp.legend()
pp.show()
```

```python
for toy in c.ll_list[5:]:
    plot_and_fit(toy, p0 = (8,5))
pp.legend(ncol=3)
pp.show()
```

We can calculate speedup given a particular error cutoff.

```python
plot_speedup_bar(c.lpt_list, gold_popt, "Length per trajectory", lpt_dir_to_x, p0=(10,3), error_val=0.5)
pp.savefig('lpt.png')
pp.show()
```

```python
plot_speedup_bar(c.ll_list, gold_popt, "Number of parallel simulations", ll_dir_to_x, width=0.2, p0=(8,5), error_val=0.5)
pp.savefig('ll.png')
```

Statistics
--------
Run identical simulation 100 times, see the difference in implied timescale.

```python
xs = c.stat_list[0].errors[:,0]
ys = np.zeros((len(xs), len(c.stat_list)))
for i in xrange(len(c2.stat_list)):
    ys[:,i] = c.stat_list[i].errors[:,1]
    
yavg = np.average(ys, axis=1)
ystd = np.std(ys, axis=1)    
```

```python
stat = LPT('stat', 1)
stat.errors = np.zeros((len(xs)-1,2))
stat.errors[:,0] = xs[1:]
stat.errors[:,1] = yavg[1:]
```

```python
pp.errorbar(xs, yavg, yerr=ystd)
plot_and_fit(stat, p0=(5, 10))
pp.ylim(ymin=0)
pp.xlabel('Walltime')
pp.ylabel('Errors')
pp.title('Error with uncertainty.')
pp.savefig('stat.png')
pp.show()
```
